```python
import pandas as pd
data = pd.read_csv('../data/Diabetes Binary Classification.csv')
data
```

```python
data.info()
```

```python
data.describe().style.background_gradient()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set(style="whitegrid")
plt.figure(figsize=(18, 18))
for i, col in enumerate(data.select_dtypes(include=['float64']).columns):
    plt.rcParams['axes.facecolor'] = 'white'
    ax = plt.subplot(4,3, i+1)
    sns.histplot(data=data, x=col, ax=ax,color='coral',kde=True)
plt.title('Data distribution of continuous variables')
plt.tight_layout()
```

```python
data.rename(columns={'Class variable (0 or 1)':'y'}, inplace=True)
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(data.corr(),annot=True,center = 0,cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(data.columns):
    plt.rcParams['axes.facecolor'] = 'white'
    ax = plt.subplot(3,4, i+1)
    sns.barplot(data=data, x='y', y=col, ax=ax,edgecolor="white",palette='viridis_r')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```

```python
data.to_csv('../data/data_1.csv',index=False)
```
