```python
import pandas as pd
data = pd.read_csv('../data/UniversalBank.csv')
data
```

```python
data.info()
```

```python
data = data.drop(['ID','ZIP Code'],axis=1)
data
```

```python
data.describe().style.background_gradient()
```

```python
data.rename(columns={'CreditCard':'y'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.distplot(data['Age'],kde=False,ax=ax1)
sns.distplot(data['Experience'],kde=False,ax=ax2,color="red")
sns.distplot(data['Income'],kde=False,ax=ax3,color="gold")

ax1.set_title('age-hist')
ax2.set_title('experience-hist')
ax3.set_title('income-hist')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.distplot(data['Family'],kde=False,ax=ax1)
sns.distplot(data['Education'],kde=False,ax=ax2,color="red")
sns.distplot(data['Mortgage'],kde=False,ax=ax3,color="gold")

ax1.set_title('family-hist')
ax2.set_title('education-hist')
ax3.set_title('mortgage-hist')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set(style="whitegrid")
plt.figure(figsize=(18, 18))
for i, col in enumerate(data.select_dtypes(include=['float64']).columns):
    plt.rcParams['axes.facecolor'] = 'white'
    ax = plt.subplot(4,3, i+1)
    sns.histplot(data=data, x=col, ax=ax,color='coral',kde=True)
plt.tight_layout()
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(data.corr(),annot=True,center = 0,cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(data.columns):
    plt.rcParams['axes.facecolor'] = 'white'
    ax = plt.subplot(3,4, i+1)
    sns.barplot(data=data, x='y', y=col, ax=ax,edgecolor="white",palette='viridis_r')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```
