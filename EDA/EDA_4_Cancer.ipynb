```python
import pandas as pd
data = pd.read_csv('../data/breast-cancer.csv')
data
```

```python
data.info()
```

```python
data.loc[data['diagnosis']=='B','y']=0
data.loc[data['diagnosis']=='M','y']=1
data['y'] = pd.to_numeric(data['y'])
data=data.drop(['diagnosis'],axis=1)
```

```python
data.describe().style.background_gradient()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize=(18,18))
sns.heatmap(data.corr(),annot=True,center = 0,cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(data.columns):
    plt.rcParams['axes.facecolor'] = 'white'
    ax = plt.subplot(4,8, i+1)
    sns.barplot(data=data, x='y', y=col, ax=ax,edgecolor="white",palette='viridis_r')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```

```python
data.to_csv('../data/data_4.csv',index=False)
```
